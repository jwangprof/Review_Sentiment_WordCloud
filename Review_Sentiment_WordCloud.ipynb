```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
```

```python
# importing dataset into a dataframe
raw = pd.read_csv("Reviews.csv")
# printing first five rows
df=raw.head(500)
```

```python
text1 = " ".join(str(Text) for Text in df['Text'].fillna(''))
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Ensure NLTK has been downloaded
nltk.download('punkt')

# Tokenize the text into words
words = word_tokenize(text1)

# Load stop words
stop_words = set(stopwords.words('english'))


# Define your custom stopwords
custom_stop_words = ['something', 'app', 'tiktok', 'tik', 'tok', 'tik tok',
                     'thing', 'good', 's', ' S ', "n ' t", "br br", "br"]

# Add custom stopwords to the set
stop_words.update(custom_stop_words)



# Filter out the stop words
filtered_text = ' '.join(word for word in words if word.lower() not in stop_words and len(word) > 3)

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a word cloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off the axis
plt.show()
```
